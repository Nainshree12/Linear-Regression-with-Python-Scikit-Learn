```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#reading dataset
url = "https://raw.githubusercontent.com/abhishek-bodapati/TSF_Data_Science_Internship_Tasks/master/To%20Explore%20Business%20Analytics/SampleSuperstore.csv"
data = pd.read_csv(url)
data.head(5)
```

```python
data.shape # Shape of the data
```

```python
data.isnull().sum() # Checks the NULL values
```

```python
data.duplicated().sum() # Checks for duplicate values
```

```python
data = data.drop_duplicates() # Drops the duplicate values
```

```python
data.info() # Information about the data
```

```python
data.shape # Shape of the data after removing duplicates
```

```python
data.describe() # Generates descriptive statistics
```

 # Exploratory Data Analysis

```python
# Correlation Matrix

corr = data.corr()
corr
```

```python
plt.figure(figsize = (12, 7))
sns.heatmap(corr, annot = True)
plt.show()
```

# Visualizing the data

```python
plt.figure(figsize = (12, 7))
sns.countplot(x = data["Ship Mode"])
plt.show()
```

```python
plt.figure(figsize = (12, 7))
sns.countplot(x = data["Segment"])
plt.show()
```

```python
plt.figure(figsize = (12, 7))
sns.countplot(x = data["Country"])
plt.show()
```

```python
plt.figure(figsize = (12, 7))
sns.countplot(x=data['City'],order=(data['City'].value_counts().head(20)).index)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (12, 7))
sns.countplot(x = data["State"])
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (12, 7))
sns.countplot(x = data["Region"])
plt.show()
```

```python
plt.figure(figsize = (12, 7))
sns.countplot(x = data["Category"])
plt.show()
```

```python
plt.figure(figsize = (12, 7))
sns.countplot(x = data["Sub-Category"])
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (12, 7))
plt.pie(data['Category'].value_counts(),labels= data['Category'].value_counts().index, startangle=90)
plt.title("Varous Categories")
plt.show()
```

```python
plt.figure(figsize = (12, 7))
sns.barplot(x = data["Sub-Category"], y = data["Profit"])
plt.xticks(rotation = 90)
```

Insights derived from the above EDA
1. Office Supplies are excellent. We have to work more on Furniture and Technology category of business.
2. Company is facing loss in Bookcases, Tables and Suppliers.
3. Improvements should be made on same day of shipment mode.
4. Profit and Discount show very weak and negative relationship. That should be kept in mind before taking any other decision related to business.
5. We have to work more in the Southern region of USA for better business.
6. There are very less people working as Copiers. So, the recruitment should be increased.
